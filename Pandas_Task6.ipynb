## scaling

```python
#norm
def norm(x):
    n = (x - x.min())/(x.max() - x.min())
    return n
```

```python
norm(df['wheel_base'])
```

```python
#standard
def stnd(x):
    n = (x - x.mean())/(x.std())
    return n
```

```python
stnd(df['price'])
```
