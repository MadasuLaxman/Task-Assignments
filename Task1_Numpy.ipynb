4. Apply Descriptive statistics to above arrays using Numpy.

```python
##measure of central Tendancy
#mean
np.mean(array1)
```

```python
##median
np.median(array1)
```

```python
##measure of Dispersion
#Range
np.max(array1)-np.min(array1)
```

```python
##variance
np.var(array1)
```

```python
##standard devition
np.std(array1)
```

```python
Q1 = np.percentile(array1,25)
Q1
```

```python
Q2 = np.percentile(array1,50)
Q2
```

```python
Q3 = np.percentile(array1,75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
Lower_limit = Q1-1.5*IQR
Lower_limit
```

```python
upper_limit = Q1+1.5*IQR
upper_limit
```

```python
##outliers
array1[array1>upper_limit]
```

```python
array1[array1<Lower_limit]
```
